```python
%%writefile HashTable.cpp
#include <iostream>
#include <functional>
#include<iomanip>//setw() \t
#include<vector>
#include <cmath>
//using namespace System;
using namespace std;

class Candidato {
	string nombre;
	float patrimonio;
	short edad;
public:
	Candidato(string nombre="",float patrimonio=50000,short edad=40) {
		this->nombre = nombre;
		this->patrimonio = patrimonio;
		this->edad = edad;
	}

	bool operator<(Candidato& c) {
		if (this->nombre.compare(c.nombre) < 0) {
			return true;
		}
		else return false;
	}

	bool operator>(Candidato& c) {
		if (this->edad> c.edad) {
			return true;
		}
		else return false;
	}

	friend ostream& operator<<(ostream& os, Candidato& c) {
		os << c.nombre << " " << c.edad << " " << c.patrimonio << endl;
		return os;
	}

};

template <class KEY,class VALUE>
class HashTable{
	//Informacion de entradas 
	struct DATA{
		KEY key;
		VALUE* value;
		DATA(KEY key,VALUE* value=new VALUE):key(key),value(value){}
	};

	vector<int> primos;
	//ATRIBUTOS
	vector <DATA*> table; 
	int capacity;
	function<int(KEY)> code;//propia de la funcion hash 
public:

    void CribaEratostenes(int n){
		primos.clear();
		bool* nroPrimos =new bool[n+1];
		//Iniciar datos
		for(int i=0;i<n;i++){
			nroPrimos[i]=true;
		}
		//Comprobacion de multiplos
		for(int i=2;i*i<=n;i++){
			for(int j=i*2;j<=n;j+=i){
				nroPrimos[j]=false;
			}
		}
        //agregar primos a nuestro arreglo nroPrimos
		for(int i=2;i<=n;i++){
			if(nroPrimos[i]) primos.push_back(i);
		}
	}
    //para asegurarnos de que no sobrepase la capacidad
    int NextPrimo(int num){
	    for(int primo: primos){
            if(primo>=num){
                return primo;
            }
        }
        return -1;
    }

	HashTable(function<int(KEY)> code,int capacity=11){
		CribaEratostenes(capacity*5.5);//generera primos
		this->code= code;
		this->capacity= NextPrimo(capacity);
		//INICIALIZAR EL ESPACIO DE LA TABLA
		for(int i=0;i<=capacity;i++){
			table.push_back(nullptr);
		}
	}

	VALUE& operator [](KEY key){
		int index = this->FunctionHash(key);
		if(table[index]==NULL){
			table[index]=new DATA(key);
		}
		return *table[index]->value;
	}

	int FunctionHash(KEY key){
		int _code=code(key);
		auto hash_generate=[=](int i){return (_code +i*i)%capacity;};// Recalcular Hash Cuando encuentra una colicion genera otro indice hash
		for (int i=0;true;i++){
			int index=hash_generate(i);
			if(table[index]==NULL || key== table[index]->key){
				return index;
			}
		}
	}

	void Print_table(){
		cout<<setw(3)<<"IDX"<<setw(15)<<"Memoria"<<setw(10)<<"LLAVE"<<setw(10)<<"VALOR"<<endl;
		for(int i=0;i<table.size();i++){
			if(table[i]!=nullptr){
				cout<<setw(3)<<i<<setw(15)<<table[i]<<setw(10)<<table[i]->key<<setw(10)<<*table[i]->value<<endl;
			}else{
				cout<<setw(3)<<i<<setw(15)<<table[i]<<setw(10)<<"..."<<setw(10)<<"..."<<endl;
			}
		}
	}

    void Show(function<void(VALUE)>c){
        for(size_t i=0;i<table.size();++i){
            if(table[i]!=nullptr){
                cout<<*table[i]->value<<endl;
            }
        }
    }

    bool buscar(KEY k, function <bool(KEY,KEY)> Com){
        int index=this->FunctionHash(k);
        for (int i=0;true;i++){
            if(Com(table[i]->key, k)) return true;
        }
        return false;
    }

};

long generateCodigoString(string word) {//crear codigo hash de string
	int suma=0;
	for(int i=0;i<word.size();i++){
		suma +=(int)word[i];
	}
	return suma;
}

long generarCodigoEntero(int i) {//crear codigo hash de Entero
	return (long)rand() % 10000;
}

long generarCodigoS(string s) {//crear codigo hash de string
	long indice = -1;
	for (long i = 0; s[i] != '\0'; ++i) {
		indice = s[i] * (long)(pow(27, i));
	}
	return indice;
}




void menu(){
    HashTable<string,int> th(generateCodigoString);
    HashTable<string,Candidato> Per(generateCodigoString);
    HashTable<int,int>Entero(generarCodigoEntero);
	cout<<"Antes de Insertar datos: "<<endl;
	th.Print_table();
	cout<<"----------------------------------"<<endl;
	
    //Insertando enteros en llaves string
    th["Mario"]=15;
	th["Pepe"]=25;
	th["Jesus"]=30;
	th["Arturo"]=17;
	th["Josefina"]=40;
	//insertando enteros en llaves int
    Entero[752096]=56;
	Entero[456789]=23;
	Entero[127895]=78;
	Entero[168894]=23;
	Entero[456789]=48;
	//Insertat personas
    Per["Tapir"]=Candidato("Pancho",1234.15,18);
	Per["Sara"]=Candidato("Aldair",90453.15,18);
	Per["Waldir"]=Candidato("Pancho",653782.15,18);
	Per["Daniel"]=Candidato("Pancho",219875.15,18);
	Per["Alicia"]=Candidato("Pancho",987632.15,18);
	cout<<Per["Tapir"];
	cout<<Per["Sara"];
	cout<<Per["Waldir"];
	cout<<Per["Daniel"];
	cout<<Per["Alicia"]<<endl;

	cout<<"Despues de insertar los datos: "<<endl;
	cout<<"-----------------------------"<<endl;
	th.Print_table();


    //Busqueda
    //th.Show([=](int v){cout<<v<<" ";});
    //th.buscar("Mario",[=](string a,string b){return a->b})
	//cout<<th["Mario"];
    //cout<<Per["Tapir"];
	//th.Print_table();
    /*int op;
    do {
       // Console::Clear()
        cout<<"\tLista Actual"<<endl;
        cout<<"\t------------"<<endl;
        cout<<"\t1. Insertar Objeto\n";
        cout<<"\t2. Busqueda x lambdas\n";
        cout<<"\t3. Eliminar por llave\n";
        cout<<"\t4. Modificar por llave\n"; 
        cout<<"\t5. salir\n";
        cout<<"-->";cin>>op;

        switch(op){

            case 1: break;

            case 2: break;

            case 3:break;

            case 4:break;
        }    

    }while(op!=13);*/
}


int main(){ 
	menu();

  cin.get();
  cin.ignore();
	return 0;

}
```
